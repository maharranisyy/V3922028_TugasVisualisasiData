```python
import pandas as pd
import matplotlib.pyplot as plt

#reading the database
data = pd.read_csv("D:/SEMESTER 2/PRAKTIK PEMROGRAMAN PYTHON/data bank.csv")

#scatter plot 
plt.scatter(data['day'], data['duration'])

#adding title
plt.title("Data")

#setting x and y labels
plt.xlabel('day')
plt.ylabel('duration')

#save to png file
plt.savefig('scatter_plot.png', dpi=500, bbox_inches='tight')

plt.colorbar()

plt.show()
```
